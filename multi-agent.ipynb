```python
import os
from crewai import Agent, Task, Crew, Process, LLM

llm = LLM(
            base_url="https://api.deepseek.com",
            api_key=os.getenv('DEEPSEEK_SK'),
            model="deepseek-chat",  # 本次使用的模型
            # temperature=0.7,  # 发散的程度
            # timeout=None,# 服务请求超时
            # max_retries=2,# 失败重试最大次数
        )

# 研究师Agent

researcher = Agent(
    role='高级研究分析师',
    goal='查找并总结 AI 最新趋势。',
    backstory='你是一名经验丰富的研究分析师，擅长发现关键趋势并整合信息。',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)
# 写Agent
writer = Agent(
    role='技术内容写作者',
    goal='根据研究结果撰写清晰易懂的博客。',
    backstory='你是一名技术写作高手，能将复杂技术转化为通俗内容。',
    verbose=True,
    allow_delegation=False,
    llm=llm,
)

# 定义任务
research_task = Task(
    description='调研 2024‑2025 年 AI 三大新兴趋势，关注实际应用与影响。',
    expected_output='详细总结三大 AI 趋势，包括要点与来源。',
    agent=researcher,
)

write_task = Task(
    description="根据研究结果撰写一篇 500 字博客，内容通俗易懂。",
    expected_output="完整的 500 字 AI 趋势博客。",
    agent=writer,
    context=[research_task],
)

# 创建团队
blog_creation_crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    process=Process.sequential,
    verbose=True,
)

# 执行
result = blog_creation_crew.kickoff()
print(result)
```
